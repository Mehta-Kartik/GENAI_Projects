Document from langchain_core.documents stores text content (page_content) with associated metadata (like source, file type, chunk ID) for RAG pipelines and text processing chains.
​

Primary Uses
RAG/Retrieval: Wraps document chunks with metadata for vector stores (e.g., {"source": "pdf1.pdf", "page": 5}).

Text Splitting: Loaders (PDFs, web pages) → split into Document objects → embeddings.

Metadata Propagation: Track origins through chains/transformers.

```python
from langchain_core.documents import Document
#page_content holds the actual text content of the document chunk, while metadata stores structured info about that content.

documents=[
    Document(
        page_content="Dogs are great companions, known for their loyalty and friendliness",
        metadata={"source":"mammal-pets-doc"},
    ),
    Document(
        page_content="Cats are independent pets that often enjoy their own space",
        metadata={"source":"mammal-pets-doc"},
    ),
    Document(
        page_content="Parrots are intelligent birds capable of mimicking human speech.",
        metadata={"source":"mammal-pets-doc"},
    ),

    Document(
        page_content= "Rabbits are social animals that need plenty of space to hop around",
        metadata={"source":"mammal-pets-doc"},
    ),
]
```

Async queries (or async operations) in LangChain use Python's asyncio for non-blocking, concurrent execution of chains/agents/LLMs, speeding up I/O-bound tasks like API calls or 
DB queries by running them in parallel instead of sequentially.


```python
import time
import asyncio
queries=["Parrot","Dog"]
print("=== SYNC Mode ===")
sync_start = time.time()
parrot_sync = vectorstore.similarity_search("Parrot", k=1)
cat_sync = vectorstore.similarity_search("Cat", k=1)
sync_time = time.time() - sync_start
print(f"Parrot sync: {parrot_sync[0].page_content[:50]}...")
print(f"Cat sync: {cat_sync[0].page_content[:50]}...")
print(f"Total sync time: {sync_time:.4f}s")
```

```python
async def async_searches():
    async_start = time.time()
    parrot_async, cat_async = await asyncio.gather(
        vectorstore.asimilarity_search("Parrot", k=1),
        vectorstore.asimilarity_search("Cat", k=1)
    )
    async_time = time.time() - async_start
    print(f"\n=== ASYNC Mode ===")
    print(f"Parrot async: {parrot_async[0].page_content[:50]}...")
    print(f"Cat async: {cat_async[0].page_content[:50]}...")
    print(f"Total async time: {async_time:.4f}s")
    print(f"Speedup: {sync_time/async_time:.1f}x")
```

```python
await async_searches()
```

# Retrievers

LangChain VectorStores do not subclass Runnable, and so cannot immediately be integrated into LCEL chains.

LangChain Retrievers **are Runnables**, so they implement a standard set of methods (e.g. synchronous and asynchronous invoke and batch operations) and are designed to be incorporated in LCEL chains.

We can create a simple version of this ourselves without subclassing Retriever. If we choose around the methods below we will build around the


```python
from typing import List
from langchain_core.documents import Document
from langchain_core.runnables import RunnableLambda
# RunnableLambda converts any Python function into a Runnable object, making it compatible with LCEL chains for seamless composition, tracing, and async/sync execution.

retriever=RunnableLambda(vectorstore.similarity_search).bind(k=1) #  k=1 extracts the top most element only
retriever.batch(["cat",'Rabbit'])
```

VectorStores implement an as_retriever method that will generate a Retriever, specifically a VectorStoreRetriever. These Retriever include specific search_type and search_kwargs attributes that identify what methods of the underlying vector to call, and how to parameterize them. For instance we can replicate the above with the followings.

```python
vectorstore.as_retriever(
    search_type="similarity",
    search_kwargs={"k":1}
)
```
