###  Message history in LLMs (via LangChain) maintains conversation context by storing past HumanMessage/AIMessage pairs, enabling follow-ups like "What did I just ask?"

#### LLMs are stateless—each call forgets prior inputs. LangChain adds chat memory using RunnableWithMessageHistory to inject history automatically

```python
# ChatMessageHistory is a simple in-memory message buffer that stores HumanMessage/AIMessage pairs for a single conversation session.
from langchain_community.chat_message_histories import ChatMessageHistory  

# BaseChatMessageHistory - Abstract interface/contract all history classes must follow. Defines methods like add_messages(), messages, clear(). You never instantiate it directly—ChatMessageHistory implements it.
from langchain_core.chat_history import BaseChatMessageHistory        
# RunnableWithMessageHistory - LCEL wrapper that automates history injection. Takes your prompt | llm | parser chain + history getter function, then magically adds conversation context to every .invoke() call.
from langchain_core.runnables.history import RunnableWithMessageHistory
```

```python
store={}
def get_session_history(session_id:str)->BaseChatMessageHistory:
    if session_id not in store:
        store[session_id]=ChatMessageHistory()
    return store[session_id]
```

```python
with_message_history=RunnableWithMessageHistory(model,get_session_history)
```

### Let's understand about managing the Conversation History.
One important concept to understand when building chatbots is how to manage conversation history. If left unmanaged the list of messages will grow unbounded and potentially overflow the context window of the LLM. Therefore it is important to add a step that limits the size of the messages you are passing in.


```python
from langchain_core.messages import HumanMessage
from langchain_core.messages import AIMessage
```

```python
from langchain_core.messages import SystemMessage,trim_messages
#trim_messages from langchain_core.messages.utils trims conversation message lists to fit within a model's token limit, preventing context window overflows during LLM calls.
trimmer = trim_messages(
    max_tokens=20,           # Keep only the last 70 tokens total
    strategy='last',         # Strategy: retain newest messages, drop oldest
    token_counter=model,     # Use model's tokenizer for accurate token counting
    include_system=True,     # Always preserve system messages if present
    allow_partial=False,     # Only trim at complete message boundaries (no partial cuts)
    start_on="human"         # When trimming from end, stop at a HumanMessage boundary
)
```

```python
messages=[
    SystemMessage(content="You'r a good AI assistant"),
    HumanMessage(content="Hi I am Jack!"),
    AIMessage(content="hi!"),
    HumanMessage(content="I like Defender car"),
    AIMessage(content="that's great"),
    HumanMessage(content="What is 4+8"),
    AIMessage(content="12"),
    HumanMessage(content="Which car is your dream car"),
    AIMessage(content="BMW M2"),
]
```
