```python
import torch
torch.cuda.is_available()
```

```python
from tensorflow.keras.preprocessing.text import one_hot
```

```python
sent=[
    'The glass of milk',
    'The glass of juice',
    'The cup of tea',
    'I am a good boy',
    'I am a good developer',
    'understand the meaning of words',
    'My videos are good'
]
```

```python
sent
```

```python
## Define the vocabulary size
voc_size=10000
```

```python
## One host representation for every word
onehotrepr=[one_hot(word,voc_size) for word in sent]
onehotrepr
```

```python
from tensorflow.keras.layers import Embedding
# from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.utils import pad_sequences
from tensorflow.keras.models import Sequential
```

```python
#pad sequences are used because our length of word is of different size and we need to make it standardized for giving as an input to the model so we would be using pad_sequences for the task

sent_length=8 #Making sequenze length
embeddingdoc=pad_sequences(onehotrepr,padding='pre',maxlen=sent_length)#padding=pre to do padding at front post for end
print(embeddingdoc)
```

```python
## Feature Representation
dim=10
```

```python
model=Sequential()
model.add(Embedding(voc_size,dim,input_length=sent_length))
model.compile('adam','mse')
```

```python
model.summary()
```

```python
model.predict(embeddingdoc)
```
