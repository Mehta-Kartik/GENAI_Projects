```python
import numpy as np
import tensorflow as tf
from tensorflow.keras.datasets import imdb
from tensorflow.keras.preprocessing import sequence
from tensorflow.keras.models import load_model
# from tensorflow.keras.layers import Embedding,SimpleRNN,Dense
```

```python

word_index=imdb.get_word_index()
reverse_word_index={value:key for key,value in word_index.items()}
```

```python
model=load_model('simple_rnn.h5')
model.save('simple_rnn.keras')
```

```python
model.summary()
```

```python
#Step2: Helper functions

def decode_review(encoded_review):
    return ' '.join([reverse_word_index.get(i-3,'?') for i in encoded_review])
def preprocess_text(text):
    words=text.lower().split()
    encoded_review=[word_index.get(word,2)+3 for word in words]
    padding_review=sequence.pad_sequences([encoded_review],maxlen=100)
    return padding_review
```

```python
##Prediction function
def predict_sentiment(review):
    preprocessed_review=preprocess_text(review)
    prediction=model.predict(preprocessed_review)

    sentiment='Positive' if prediction[0][0]>.5 else 'Negative'
    return sentiment,prediction[0][0]
```

```python
example_review="This movie was fantastic! The acting was great and plot was thrilling"
```

```python
predict_sentiment(example_review)
```
