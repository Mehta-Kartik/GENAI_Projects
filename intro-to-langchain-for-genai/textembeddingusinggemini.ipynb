```python
## We can also set custom dimension as below
from langchain_google_genai import GoogleGenerativeAIEmbeddings
key=os.getenv('GEMINI')
embedding_11=GoogleGenerativeAIEmbeddings(model="gemini-embedding-001",google_api_key=key,dimensions=120)
```

```python
from langchain_community.document_loaders import TextLoader
loader=TextLoader("speech.txt")
docs=loader.load()
```

```python
docs
```

```python
from langchain_text_splitters import RecursiveCharacterTextSplitter

text_splitter=RecursiveCharacterTextSplitter(chunk_size=100,chunk_overlap=20)
findoc=text_splitter.split_documents(docs)
findoc
```

```python
## Now I will try to combine vector embedding and vector store
from langchain_community.vectorstores import Chroma
db=Chroma.from_documents(findoc,embedding)
```
