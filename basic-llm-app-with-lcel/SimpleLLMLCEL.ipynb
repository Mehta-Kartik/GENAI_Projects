```python
import os
from dotenv import load_dotenv
load_dotenv()


groq_api_key=os.getenv("GROQ_API_KEY")
# groq_api_key
```

```python
from langchain_groq import ChatGroq
from langchain_google_genai import ChatGoogleGenerativeAI
```

```python
model=ChatGroq(model="openai/gpt-oss-120b",groq_api_key=groq_api_key)
```

```python
from langchain_core.messages import HumanMessage,SystemMessage

messages=[
    SystemMessage(content="Translate the following from english to hindi"),
    HumanMessage(content="How are you?")
]


response=model.invoke(messages)
```

```python
from langchain_core.output_parsers import StrOutputParser
parser=StrOutputParser()
parser.invoke(response)
```

```python
#Using LCEL we can chain components
chain=model|parser
chain.invoke(messages)
```

```python
##We can also use prompt Template to get our solution
from langchain_core.prompts import ChatPromptTemplate

generic_template="Translate the following into {language}"
prompt=ChatPromptTemplate.from_messages(
    [("system",generic_template),("user","{text}")]
)

```

```python
result=prompt.invoke({"language":"Hindi","text":"Delhi is the capital of India"})
```

```python
chain=prompt|model|parser
result=chain.invoke({"language":"Hindi","text":"Delhi is the capital of India"})
```
