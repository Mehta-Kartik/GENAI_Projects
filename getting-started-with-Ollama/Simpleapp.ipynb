```python
## Let's create a simple GENAI App using Langchain
import os
from dotenv import load_dotenv
load_dotenv()
os.environ["GEMINI_API_KEY"]    =os.getenv("GEMINI")
os.environ["LANGCHAIN_API_KEY"]=os.getenv("LANGCHAIN_API_KEY")
os.environ["LANGCHAIN_TRACING_V2"]="true"
# enables LangSmith's v2 tracing system for monitoring and debugging LangChain applications.
os.environ["LANGCHAIN_PROJECT"]=os.getenv("LANGCHAIN_PROJECT")
```

```python
## Data Ingestion
from langchain_community.document_loaders import WebBaseLoader
```

```python
loader=WebBaseLoader(("https://docs.langchain.com/langsmith/observability-quickstart#tracing-quickstart"))
loader
```

```python
docs=loader.load()
```

```python
print(docs)
```

```python
### Loading the data now time for diving them into chunks


from langchain_text_splitters import RecursiveCharacterTextSplitter
split=RecursiveCharacterTextSplitter(chunk_size=240,chunk_overlap=40)
docd =split.split_documents(docs
                      )
```

```python
docd 
```

```python
from langchain_google_genai import GoogleGenerativeAIEmbeddings
embeddings=GoogleGenerativeAIEmbeddings(model="gemini-embedding-001")
```

```python
from langchain_community.vectorstores import FAISS
vectorsdb=FAISS.from_documents(docd,embeddings)
```

```python
vectorsdb
```

```python
query="Observability is a critical requirement for applications built"
result=vectorsdb.similarity_search(query)
result
```

```python
query2="LangSmith has two usage limits: total traces and Extended"
result2=vectorsdb.similarity_search(query2)
print(result2[0].page_content)
```

```python
from langchain_google_genai import ChatGoogleGenerativeAI
llm=ChatGoogleGenerativeAI(model="gemini-2.5-flash")
```

```python
## Retrieval Chain, Documents chain
from langchain_classic.chains.combine_documents import create_stuff_documents_chain
from langchain_core.prompts import ChatPromptTemplate

prompt=ChatPromptTemplate.from_template(
    """
    Answer the following question based only on the provided context:
    <context>
    {context}
    </context>

"""
)
document_chain=create_stuff_documents_chain(llm,prompt)
```

```python
print("Jay Ganesh")
```

```python
document_chain
```

```python
from langchain_core.documents import Document
document_chain.invoke({
    "context": [Document(page_content="Observability is a critical requirement...")],
    "user": "Observability is a critical requirement for applications built"
    
})  
```

```python
document_chain.invoke({
    "context": [Document(page_content="Observability is a critical requirement for applications built with Large Language Models (LLMs). LLMs are non-deterministic, which means that the same prompt can produce different responses.")],
    "question": "What is observability?"  # Clear Q
})
```

```python
vectorsdb
```

```python
retriever=vectorsdb.as_retriever()
```

```python
from langchain_classic.chains import create_retrieval_chain
retriverchain=create_retrieval_chain(retriever,document_chain)
```

```python
response=retriverchain.invoke({"input":"What is observability"})
```
